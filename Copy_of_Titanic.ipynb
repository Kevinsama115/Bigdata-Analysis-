```python
##타이타닉
import pandas as pd

# GitHub의 Raw 파일 URL (웹에서 직접 읽기 위해 raw.githubusercontent.com 주소 사용)
df=pd.read_csv("https://raw.githubusercontent.com/ai7dnn/2026-1-BDA/main/titanic1309.csv",index_col=0)
df
```

```python
#df.corr()
#0일수록 약하고 1,-1강하다

import pandas as pd
import matplotlib.pyplot as plt

# 데이터 불러오기


# 수치형 컬럼만 선택
cols = ['Survived', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare']
corr = df[cols].corr()

# 히트맵 그리기
plt.figure(figsize=(8, 6))
plt.imshow(corr, cmap='RdYlGn', vmin=-1, vmax=1)
plt.colorbar()

# 축 라벨
plt.xticks(range(len(cols)), cols, rotation=45)
plt.yticks(range(len(cols)), cols)

# 각 칸에 숫자 표시
for i in range(len(cols)):
    for j in range(len(cols)):
        plt.text(j, i, f'{corr.values[i, j]:.2f}',
                 ha='center', va='center', fontsize=10)

plt.title('Correlation Heatmap (생존여부 상관관계)')
plt.tight_layout()
plt.savefig('corr_heatmap.png', dpi=150)
plt.show()
```

변수상관계수해석Pclass-0.34등급 낮을수록(3등석) 사망 확률↑Fare+0.26운임 높을수록 생존 확률↑Age-0.08약한 음의 상관 (고령일수록 약간 불리)Parch+0.08거의 무관SibSp-0.04거의 무관

```python
# 사망자만
dead = df[df['Survived'] == 0]

# 항구별 색상
colors = {'C': 'red', 'Q': 'blue', 'S': 'green'}

# 산점도
plt.figure(figsize=(8, 5))

for port in ['C', 'Q', 'S']:
    data = dead[dead['Embarked'] == port]
    plt.scatter(data['Age'], data['Fare'],
                label=port,
                color=colors[port],
                alpha=0.5)

plt.title('승선항구별 사망자 (Age vs Fare)')
plt.xlabel('나이')
plt.ylabel('운임')
plt.legend(title='항구')

plt.show()
```

```python
# df_Cabin=df.Cabin.value_counts()
# print(df_Cabin)
# plt.bar(df_Cabin.index,df_Cabin)
# plt.figure(figsize=(19,8))

import pandas as pd
import matplotlib.pyplot as plt



# Cabin 빈도 계산
df_Cabin = df['Cabin'].value_counts()
print(df_Cabin)

# 시각화 (figsize는 plt.bar 이전에 선언)
plt.figure(figsize=(19, 8))
plt.bar(df_Cabin.index, df_Cabin, color='steelblue', edgecolor='white', linewidth=0.5)

plt.title('Cabin 별 승객 수', fontsize=16, fontweight='bold', pad=15)
plt.xlabel('Cabin', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(rotation=90, fontsize=8)   # x축 라벨 90도 회전 (겹침 방지)
plt.grid(axis='y', alpha=0.4)
plt.tight_layout()                    # 라벨 잘림 방지
plt.show()
```
